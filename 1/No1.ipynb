##### 1.4.2 类
##### 如果用户自己定义类的话 就可以自己创建数据类型 此外也可以定义原创的方法和属性

##### class 类名：
            def __init__(self,参数，...)
            def 方法1(self,参数)
这里有一个特殊的__init__方法 是进行初始化的方法 也称为构造方法只在生成类的实例的时候被调用一次 注意**所有方法**需要传入自身self

```python
class Man:
    def __init__(self,name):
        self.name=name
        print("Initialized!")
    def hello(self):
        print("Hello "+self.name+"!")
    def goodbye(self):
        print("Goodbye "+self.name+"!")
m = Man("David")
m.hello()
m.goodbye()
m.name = "Lili"
m.hello()
m.goodbye()
```

```python
#1.5 numpy
import numpy as np
```

```python
#1.5.2 生成numpy数组
x = np.array([1,2,3,4,5])
print(x)
print(type(x))
```

```python
#1.5.3 numpy的算术运算
x = np.array([1, 2, 3])
y = np.array([2, 4, 6])
z = x + y
print(z)
z = x - y
print(z)
z = x * y
print(z)
z = x / y
print(z)

#这里x和y数组元素的个数必须相同，不仅可以和元素个数相同的进行运算 也可以和单一的数值组合起来进行运算 这个就是np的广播机制
```

```python
#1.5.4 numpy的N维数组
A = np.array([[1, 2], [3, 4]])
print(A)
print(A.shape)
print(A.dtype)#查看矩阵元素的数据类型
B = np.array([[3, 0], [0, 6]])
C = A + B
print(C)
C = A * B
print(C)
```

```python
#1.5.5 广播
#np中形状不同的数组之间也可以运算 这个巧妙的功能成为广播
# 就是把小数组扩展
A = np.array([[1, 2], [3, 4]])
B = 10
C = A + B
print(C)
B = np.array([[10,10]])
C = A * B
print(C)
C = B * A
print(C)
```

```python
# 1.5.6 访问元素
X = np.array([[1,2], [3,4]])
print(X)
print(X[0])
print(X[1][1])
for row in X:
    for num in row:
        print(num)

#除此之外还可以使用数组访问各个元素
X = X.flatten()#把X转换成为1维数组
print(X)
print(X[np.array([1,2])])
print(X > 2)
print(X[X > 2])
```

```python
#1.6 matplotlib
import matplotlib.pyplot as plt
X = np.arange(0, 6, 0.1)#以 0.1为单位生成0～6的数据
Y = np.sin(X)
plt.plot(X, Y)
plt.show()
```

```python

X = np.arange(0, 6, 0.1)
y1 = np.sin(X)
y2 = np.cos(X)
plt.plot(X, y1, label="sin")
plt.plot(X, y2, linestyle = "--", label="cos")
plt.xlabel("x")
plt.ylabel("y")
plt.title('sin & cos')
plt.legend()
plt.show()
```
