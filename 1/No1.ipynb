```python
#1.3.4列表
a = [1,2,3,4,5]
print(a)
#获取列表长度
len(a)
#访问第一个元素值
print(a[0])
a[4] = 99 #赋值
#切片的方式访问 注意是左闭右开的
print(a[0:2])
print(a[1:])#获取从索引为1的元素到最后一个元素
print(a[:3])#获取从索引为0的元素（也就是第一个元素）到3（不包含索引为3的）
#获取从第一个元素到最后一个元素之前的一个元素
print(a[:-1])
```

**#获取从第一个元素到最后一个元素之前的一个元素**

print(a[:-1])

###### 索引-1代表最后一个元素 -2对应倒数第二个元素 由于是左闭右开所以是取不到对应的位置的
