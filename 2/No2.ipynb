##### 感知机
感知机是神经网络的起源  感知机接受多个输入信号 输出一个信号 感知机只有（1/0）两种取值，0代表不传递信号 1代表传递信号

具体的计算流程是：输入信号x1 x2 会被分别乘以固定的权重(w1x1, w2x2)
然后神经元会计算传送过来的信号的总和 只有当这个总和超过了某个限值时，才会输出1 也就是被激活 这里将这个限值成为阈值
感知机的不同输入信号对应着不同的权重 权重越大意味着这个信号越重要

#####  2，2 简单的逻辑电路
##### 与或非门 书里简单介绍了怎么用感知机的取值来实现与或非门 取值并不唯一
#### 这里我们是人类决定了取值 但是机器学习课程则是把决定权交给了机器自动进行，学习就是确定合适参数的过程
相同构造的感知机 只需要适当的调整参数的值 就可以像变色龙演员一样表演不同的角色

```python
import numpy as np


#2.3.1 感知机的简单实现
def AND(x1, x2):
    w1, w2, theta = 0.5, 0.5, 0.7
    tmp = w1 * x1 + w2 * x2
    if tmp <= theta:
        return 0
    else:
        return 1
print(AND(0, 0))
print(AND(0, 1))
print(AND(1, 0))
print(AND(1, 1))
```

```python
import numpy as np
x = np.array([0, 1])
w = np.array([0.5, 0.5])
b = -0.7
print(w * x)
np.sum(w * x)
print(np.sum(w * x) + b)
```

```python
#2.3.3 使用权重和偏值实现与门
def AND (x1, x2):
    x = np.array([x1, x2])
    w = np.array([0.5, 0.5])
    b = -0.7
    tmp = np.sum(x * w) + b
    if tmp <= 0:
        return 0
    else:
        return 1
print(AND(0, 0))
print(AND(0, 1))
print(AND(1, 0))
print(AND(1, 1))
```

##### 注意偏值和权重的作用是不一样的 权重决定了信号的重要程度 但是偏值决定的是神经元被激活的容易程度 有时候也会把偏值和权重统称为权重或者参数

```python
#与非门
def NAND(x1 , x2):
    x = np.array([x1, x2])
    w = np.array([-0.5, -0.5])
    b = 0.7
    tmp = np.sum(x * w) + b
    if tmp <= 0:
        return 0
    else:
        return 1
def OR(x1, x2):
    x = np.array([x1, x2])
    w = np.array([0.5, 0.5])
    b = - 0.2
    tmp = np.sum(x * w) + b
    if tmp <= 0:
        return 0
    else:
        return 1
```

###### 感知机的局限性：没办法实现异或门
所以需要组合来实现非线性的表示 也就有了多层感知机


```python
# 实现异或门
def XOR(x1, x2):
    s1 = NAND(x1, x2)
    s2 = OR(x1, x2)
    return AND(s1, s2)
print(XOR(0, 0))
print(XOR(0, 1))
print(XOR(1, 0))
print(XOR(1, 1))
```
